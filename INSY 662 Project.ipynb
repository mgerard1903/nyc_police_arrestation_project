## Demographic Disparities

```python
racial_outcomes = sqf_data.groupby('SUSPECT_RACE_DESCRIPTION')[binary_columns].mean() * 100  # Convert to percentages

# Calculate unjustified rates
racial_outcomes['Unjustified Stop Rate'] = 100 - racial_outcomes['SUSPECT_ARRESTED_FLAG']
racial_outcomes['Unjustified Frisk Rate'] = racial_outcomes['FRISKED_FLAG'] * (1 - racial_outcomes['SUSPECT_ARRESTED_FLAG'] / 100)
racial_outcomes['Unjustified Search Rate'] = racial_outcomes['SEARCHED_FLAG'] * (1 - racial_outcomes['SUSPECT_ARRESTED_FLAG'] / 100)

# Plotting the bar chart with slightly reduced dimensions and adjusted margins
fig, ax = plt.subplots(figsize=(16, 8))

# Set bar width and adjust margins to fill the plot area
racial_outcomes[['Unjustified Stop Rate', 'Unjustified Frisk Rate', 'Unjustified Search Rate']].plot(kind='bar', ax=ax, width=0.75, legend=False)
plt.title('Racial Disparities in Unjustified Stop, Frisk, and Search Rates')
plt.xlabel('Suspect Race')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=80)

# Place legend outside the plot area
plt.legend(['Unjustified Stop Rate', 'Unjustified Frisk Rate', 'Unjustified Search Rate'],
           title='Outcome Type', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust margins to make bars larger within the plot area
plt.subplots_adjust(left=0.15, right=0.8, top=0.9, bottom=0.3)

plt.show()
```

```python

# Step 1: Create a copy of the original data and define income bins on this copy
sqf_data_copy = sqf_data.copy()
income_bins = [0, 20000, 40000, 60000, 80000]
income_labels = ['$0-$20k', '$20k-$40k', '$40k-$60k', '$60k-$80k']
sqf_data_copy['Income_Level'] = pd.cut(sqf_data_copy['Median_Income'], bins=income_bins, labels=income_labels)

# Step 2: Calculate Unjustified Arrest and Frisk Rates on the copied data
# Define new columns for unjustified arrest and frisk
sqf_data_copy['Unjustified_Arrest'] = sqf_data_copy['SUSPECT_ARRESTED_FLAG'].apply(lambda x: 0 if x == 1 else 1)  # 1 if no arrest, 0 if arrested
sqf_data_copy['Unjustified_Frisk'] = sqf_data_copy['FRISKED_FLAG'] * sqf_data_copy['Unjustified_Arrest']  # Frisk without arrest

# Step 3: Group by income level and race, then calculate mean rates for unjustified outcomes
unjustified_columns = ['Unjustified_Arrest', 'Unjustified_Frisk']
income_race_unjustified_outcomes = sqf_data_copy.groupby(['Income_Level', 'SUSPECT_RACE_DESCRIPTION'])[unjustified_columns].mean() * 100  # Convert to percentages
income_race_unjustified_outcomes = income_race_unjustified_outcomes.rename(columns={
    'Unjustified_Arrest': 'Unjustified Arrest Rate (%)',
    'Unjustified_Frisk': 'Unjustified Frisk Rate (%)'
}).reset_index()

# Step 4: Visualize unjustified outcomes by income level and race

# Unjustified Arrest Rate by Income Level and Race
plt.figure(figsize=(14, 10))
sns.barplot(data=income_race_unjustified_outcomes, x='Income_Level', y='Unjustified Arrest Rate (%)', hue='SUSPECT_RACE_DESCRIPTION')
plt.title('Unjustified Arrest Rate by Income Level and Race')
plt.xlabel('Income Level')
plt.ylabel('Unjustified Arrest Rate (%)')
plt.xticks(rotation=45)
plt.legend(title='Race')
plt.tight_layout()
plt.show()

# Unjustified Frisk Rate by Income Level and Race
plt.figure(figsize=(14, 10))
sns.barplot(data=income_race_unjustified_outcomes, x='Income_Level', y='Unjustified Frisk Rate (%)', hue='SUSPECT_RACE_DESCRIPTION')
plt.title('Unjustified Frisk Rate by Income Level and Race')
plt.xlabel('Income Level')
plt.ylabel('Unjustified Frisk Rate (%)')
plt.xticks(rotation=45)
plt.legend(title='Race')
plt.tight_layout()
plt.show()
```

```python
# Step 1: Filter for unjustified stops (where no arrest was made)
unjustified_stops = sqf_data[sqf_data['SUSPECT_ARRESTED_FLAG'] == 0]

# Step 2: Group by race and suspected crime description, and count occurrences
crime_by_race = unjustified_stops.groupby(['SUSPECT_RACE_DESCRIPTION', 'SUSPECTED_CRIME_DESCRIPTION']).size().reset_index(name='Unjustified_Count')

# Step 3: Identify the most common unjustified crime per race
# Sort within each race by the count in descending order, then take the top entry per race
most_common_unjustified_crime_by_race = crime_by_race.sort_values(['SUSPECT_RACE_DESCRIPTION', 'Unjustified_Count'], ascending=[True, False]) \
    .groupby('SUSPECT_RACE_DESCRIPTION').head(1)

# Display the result
print(most_common_unjustified_crime_by_race)
```

For nearly all racial groups, Criminal Possession of a Weapon (CPW) appears as the predominant suspected crime that does not result in an arrest, suggesting a consistent pattern where CPW-related stops often lack sufficient grounds to proceed to an arrest. The notably high counts among Black and White Hispanic groups, in particular, may indicate a focus on CPW suspicions within these communities, which often dont lead to arrest outcomes. This pattern could reflect potential biases or systematic differences in how CPW suspicions are applied across different racial groups.

## Multicollinearity

```python
correlation_matrix = sqf_model_data_dummified.corr()

# Step 2: Extract the upper triangle of the correlation matrix
# This avoids duplicate pairs and self-correlations
upper_triangle = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))

# Step 3: Unstack, take the absolute value, and filter for correlations >= 0.8
high_corr_pairs = upper_triangle.stack().reset_index()
high_corr_pairs.columns = ['Variable 1', 'Variable 2', 'Correlation']
high_corr_pairs = high_corr_pairs[high_corr_pairs['Correlation'].abs() >= 0.8]

# Display the result
print(high_corr_pairs)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Step 1: Convert boolean columns (True/False) to integer (1/0)
sqf_model_data_vif = sqf_model_data_dummified.copy()  # Make a copy to avoid altering the original data
sqf_model_data_vif = sqf_model_data_vif.apply(lambda col: col.astype(int) if col.dtype == 'bool' else col)

# Step 2: Select all relevant features (numeric, integer, float, and converted boolean types)
# Exclude the target variable if applicable
features = sqf_model_data_vif.select_dtypes(include=['float64', 'int64', 'int32'])

# Step 3: Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data['Feature'] = features.columns
vif_data['VIF'] = [variance_inflation_factor(features.values, i) for i in range(features.shape[1])]
```

## XGBOOST

### XGBOOST Model Definition and hyperparameter tuning

```python
from xgboost import XGBClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, log_loss

# Set up the parameter grid for XGBoost
param_grid = {
    'estimator__n_estimators': [50, 100, 150],         # Number of trees
    'estimator__max_depth': [3, 5, 7],                 # Maximum depth of each tree
    'estimator__learning_rate': [0.01, 0.1, 0.2],      # Learning rate
    'estimator__subsample': [0.7, 0.8, 1.0],           # Fraction of samples per tree
    'estimator__colsample_bytree': [0.7, 0.8, 1.0]     # Fraction of features per tree
}

y = sqf_model_data_final[[ 'Unjustified_Frisk']]

# Initialize the base model
xgb = XGBClassifier(random_state=42, eval_metric='logloss')

# Wrap the model in a MultiOutputClassifier to handle multiple outputs
multi_xgb = MultiOutputClassifier(xgb)

# Define the custom scoring function for log loss
log_loss_scorer = make_scorer(log_loss, greater_is_better=False, needs_proba=True)

# Set up the grid search with log loss as the scoring metric
grid_search = GridSearchCV(multi_xgb, param_grid, cv=3, scoring=log_loss_scorer, n_jobs=-1, verbose=2)
grid_search.fit(X, y)

# Get the best parameters
print("Best parameters found: ", grid_search.best_params_)
best_model = grid_search.best_estimator_
```

### XGBoost Model Performance

```python
from sklearn.model_selection import cross_val_score
cv_log_loss_scores = cross_val_score(best_model, X, y, cv=3, scoring=log_loss_scorer, n_jobs=-1)

# Print the log loss for each fold and the mean log loss
print("Cross-validation Log Loss scores for each fold:", -cv_log_loss_scores)  # Convert to positive values for readability
print("Mean CV Log Loss:", -np.mean(cv_log_loss_scores))
```
